```python
import pandas as pd
from pathlib import Path


school_data_path = Path("Resources/schools_complete.csv")
student_data_path = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_path)
student_data = pd.read_csv(student_data_path)

# Combine the data into a single dataset.  
school_data_total = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_total.head()
```

## District Summary

```python
 # Calculate the total number of unique schools
school_count = len(school_data_total["school_name"].unique())
print(school_count)
```

```python
# Calculate the total number of students
student_count = len(school_data_total["student_name"])
print(student_count)
```

```python
# Calculate the total budget
total_budget = school_data["budget"].sum()
print(total_budget)
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_total["math_score"].mean()
print(average_math_score)
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_total["reading_score"].mean()
print(average_reading_score)
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_total[(school_data_total["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
print(passing_math_percentage)
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_total[(school_data_total["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
print(passing_reading_percentage)
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_total[
    (school_data_total["math_score"] >= 70) & (school_data_total["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
print(overall_passing_rate)
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
data = {"Total Schools":[school_count],"Total Students":[student_count],
        "Total Budget":[total_budget],"Average Math Score":[average_math_score],
        "Average Reading Score":[average_reading_score],"% Passing Math":[passing_math_percentage],
        "% Passing Reading":[passing_reading_percentage],"% Overall Passing":[overall_passing_rate]}

district_summary = pd.DataFrame(data)

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary.head()
```

## School Summary

```python
# show school names
df = school_data.groupby("school_name").count().reset_index()
df[["school_name"]].head()
```

```python
#show school types
grp = school_data_total.groupby(["school_name"])
school_types = pd.Series(grp.first()["type"])
school_types.head().to_frame("School Type")
```

```python
#total student count per school
per_school_counts = school_data_total.groupby("school_name").size()
per_school_counts.head().map("{:,}".format).to_frame("Student_Count")
```

```python
# total school budget
grp = school_data_total.groupby("school_name")
per_school_budget = pd.Series(grp.first()[["budget"]].squeeze())
per_school_budget.head().map("${:,.2f}".format).to_frame("School Budget")
```

```python
# per student budget
df = school_data_total.groupby("school_name")
per_student_budget = (df["budget"].mean() / df["student_name"].count())
per_student_budget.head().map("${:,.2f}".format).to_frame("Per Student Budget")
```

```python
# average math score
Average_Math = school_data_total.groupby("school_name")["math_score"].mean()
Average_Math.head().map("{:.2f}".format).to_frame("Average Math Score")
```

```python
# average reading score
Average_Reading = school_data_total.groupby("school_name")["reading_score"].mean()
Average_Reading.head().map("{:.2f}".format).to_frame("Average Reading Score")
```

```python
# number of students who passed math and reading
df = school_data_total
passed_math_all = df.loc[(df["math_score"]) >=70]
passed_reading_all = df.loc[(df["reading_score"]) >=70]
passed_overall_all = df.loc[(df["math_score"] >=70) & (df["reading_score"]>=70)]
```

```python
#  percentage of students who passed math per school
passed_math = passed_math_all.groupby("school_name").count()["student_name"] / per_school_counts * 100
passed_math.head().map("{:.2f}%".format).to_frame("Passed Math")
# grp_school_data = school_data_total.groupby("school_name")
# grp_students = grp_school_data["student_name"].count()s
# passed_math = school_data_total.query("math_score >= 70")["student_name"].count() / grp_students
#per_passed_math 
```

```python
# # percentage of students who passed reading
# grp_school_data = school_data_total.groupby("school_name")
# grp_students = grp_school_data["student_name"].count()
# passed_reading = school_data_total.query("reading_score >= 70")["student_name"].count() / grp_students
# passed_reading.head().map("{:.2f}%".format).to_frame("Passed Reading")
passed_reading = passed_reading_all.groupby("school_name").count()["student_name"] / per_school_counts * 100
passed_reading.head().map("{:.2f}%".format).to_frame("Passed Reading")
```

```python
# percentage of overall students who passed math and reading
passed_overall = ((Average_Math + Average_Reading)/2)
passed_overall.head().map("{:.2f}%".format).to_frame("Passed Overall")
```

```python
# create a school summary dataframe
dfdict = {"School Type":school_types,
          "Total Students":per_school_counts,
          "Total School Budget":per_school_budget,
          "Per Student Budget":per_student_budget,
          "Average Math Score":Average_Math,
          "Average Reading Score":Average_Reading,
          "% Passing Math":passed_math,
          "% Passing Reading":passed_reading,
          "% Overall Passing":passed_overall
          }

per_school_summary = pd.DataFrame(dfdict)
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

per_school_summary.index_name = None
per_school_summary.head()
```

# Math Scores by Grade

```python
# average students math scores by grade per school
df = school_data_total
grade_nine = df[(df["grade"] == "9th")]
grade_ten = df[(df["grade"] == "10th")]
grade_eleven = df[(df["grade"] == "11th")]
grade_twelve = df[(df["grade"] == "12th")]

scoretype = "math_score"

scores_nine = grade_nine.groupby("school_name")[scoretype].mean()
scores_ten = grade_ten.groupby("school_name")[scoretype].mean()
scores_eleven = grade_eleven.groupby("school_name")[scoretype].mean()
scores_twelve = grade_twelve.groupby("school_name")[scoretype].mean()

math_scores_by_grade = pd.DataFrame({"9th":scores_nine,"10th":scores_ten,
                                     "11th":scores_eleven,"12th":scores_twelve})

math_scores_by_grade.index.name = None
math_scores_by_grade.head()
```

# Reading Scores by Grade

```python
# average students reading scores by grade per school
df = school_data_total
grade_nine = df[(df["grade"] == "9th")]
grade_ten = df[(df["grade"] == "10th")]
grade_eleven = df[(df["grade"] == "11th")]
grade_twelve = df[(df["grade"] == "12th")]

scoretype = "reading_score"

scores_nine = grade_nine.groupby("school_name")[scoretype].mean()
scores_ten = grade_ten.groupby("school_name")[scoretype].mean()
scores_eleven = grade_eleven.groupby("school_name")[scoretype].mean()
scores_twelve = grade_twelve.groupby("school_name")[scoretype].mean()

reading_scores_by_grade = pd.DataFrame({"9th":scores_nine,"10th":scores_ten,
                                     "11th":scores_eleven,"12th":scores_twelve})

reading_scores_by_grade.index.name = None
reading_scores_by_grade.head()
```

# Scores by School Spending

```python
# create the bins
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# copy per_school_summary
school_summary_df = per_school_summary.copy()
```

```python
# Use 'pd_cut' to categorize spending based on the bins
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=spending_labels)

spending_math_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

spending_summary = pd.DataFrame({"Average Math Score":spending_math_scores,
                                 "Average Reading Score":spending_reading_scores,
                                 "% Passing Math":spending_passing_math,
                                 "% Passing Reading":spending_passing_reading,
                                 "% Overall Passing":overall_passing_spending})
spending_summary
```

# Score by School Size

```python
# create the bins
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium(1000-2000)", "Large (2000-5000)"]
```

```python
# Use 'pd_cut' to categorize spending based on the bins
per_school_summary["Total Students"] = pd.cut(per_school_counts, size_bins, labels=size_labels)

size_math_scores = per_school_summary.groupby(["Total Students"])["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby(["Total Students"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["Total Students"])["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby(["Total Students"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["Total Students"])["% Overall Passing"].mean()

size_summary = pd.DataFrame({"Average Math Score":size_math_scores,
                             "Average Reading Score": size_reading_scores,
                             "% Passing Math":size_passing_math,
                             "% Passing Reading":size_passing_reading,
                             "% Overall Passing":size_overall_passing})

size_summary
```
